```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

caminho_arquivo = '/content/drive/MyDrive/Colab/CHURN_TELECON_MOD08_TAREFA.csv'

df = pd.read_csv(caminho_arquivo, delimiter = ';')
df.head(10)

# Verificando tipos de dados
print(df.info())

# Exibir dados pr√©-processados
print(df.head())

# 2) A) DADOS FALTANTES
dados_faltantes = df.isnull().sum()
porcentagem_faltantes = (df.isnull().sum() / len(df)) * 100
df_faltantes = pd.DataFrame({'Quantidade Faltante': dados_faltantes,
                             'Porcentagem (%)': porcentagem_faltantes})

# Exibindo apenas as colunas com dados faltantes
df_faltantes = df_faltantes[df_faltantes['Quantidade Faltante'] > 0]

print("--- Dados Faltantes e suas Percentagens ---")
print(df_faltantes)

#2) B) Excluindo os faltantes apenas em 'Genero' e 'Churn' pois s√£o uma quantidade muito pequena de dados
df_cleaned = df.dropna(subset=['Genero', 'Churn'])
print(df_cleaned[['Genero', 'Churn']].isnull().sum())
df_cleaned.info()

#2) C) Para 'PhoneService' preenchemos com a categoria mais frequente
moda_phone = df_cleaned['PhoneService'].mode()[0]
df_cleaned['PhoneService'] = df_cleaned['PhoneService'].fillna(moda_phone)

# Para 'Pagamento_Mensal' preenchemos com a mediana
mediana_pag = df_cleaned['Pagamento_Mensal'].median()

print("Mediana:", mediana_pag)
df_cleaned['Pagamento_Mensal'] = df_cleaned['Pagamento_Mensal'].fillna(mediana_pag)
print(df_cleaned[['PhoneService', 'Pagamento_Mensal']].isnull().sum())

#3)
colunas_categoricas = [
    'Genero', 'Casado', 'Dependents', 'PhoneService',
    'Servico_Internet', 'Servico_Seguranca', 'Suporte_Tecnico',
    'StreamingTV', 'Tipo_Contrato', 'PaymentMethod', 'Churn'
]

for col in colunas_categoricas:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip().str.lower()

df_cleaned['Genero'] = df_cleaned['Genero'].replace({
    'm': 'masculino',
    'masc': 'masculino',
    'masculino': 'masculino',
    'f': 'feminino',
    'fem': 'feminino',
    'feminino': 'feminino'
})

df_cleaned['Casado'] = df_cleaned['Casado'].replace({
    'yes': 'sim',
    'y': 'sim',
    'no': 'nao',
    'n': 'nao'
})

#Conferindo valores √∫nicos para identificar se ainda existem inconsist√™ncias
for col in colunas_categoricas:
    print(f"{col}: {df_cleaned[col].unique()}")
```
